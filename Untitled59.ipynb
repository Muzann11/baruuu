```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/download11.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis("off")
plt.title("Original")
plt.show()
```

#smoothing

```python
gauss = cv2.GaussianBlur(img_rgb, (5,5), 0)
median = cv2.medianBlur(img_rgb, 5)
mean = cv2.blur(img_rgb, (5, 5))

plt.subplot(1,3,1); plt.imshow(gauss); plt.title("Gaussian"); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(median); plt.title("Median"); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(mean); plt.title("Mean"); plt.axis("off")
plt.show()
```

#sharpening

```python
mawar = cv2.imread("/content/mawar.png")
mawar_rgb = cv2.cvtColor(mawar, cv2.COLOR_BGR2RGB)
mawar_gray = cv2.cvtColor(mawar_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(mawar_gray, cmap="gray")
plt.title("Original")
plt.axis("off")
plt.show()
```

```python
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
sharpen = cv2.filter2D(mawar_gray, -1, kernel)

plt.imshow(sharpen, cmap="gray")
plt.title("Sharpened")
plt.axis("off")
plt.show()
```

```python
# Lowpass pakai Gaussian blur
lowpass = cv2.GaussianBlur(mawar_gray, (5,5), 1)

plt.imshow(lowpass, cmap="gray")
plt.title("Lowpass (Blur)")
plt.axis("off")
plt.show()
```

```python
# Highpass = Original - Lowpass
highpass = cv2.subtract(mawar_gray, lowpass)

plt.imshow(highpass, cmap="gray")
plt.title("Highpass (Detail)")
plt.axis("off")
plt.show()
```

```python
sharp = cv2.add(mawar_gray, highpass)

plt.imshow(sharp, cmap="gray")
plt.title("Sharpened Image")
plt.axis("off")
plt.show()
```

#segmenatation

```python
gray = cv2.cvtColor(mawar_rgb, cv2.COLOR_RGB2GRAY)
_, th = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.imshow(th, cmap="gray")
plt.title("Otsu Threshold")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(mawar_rgb, cv2.COLOR_RGB2GRAY)
th = cv2.adaptiveThreshold(gray, 255,
                           cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                           cv2.THRESH_BINARY, 11, 2)
plt.imshow(th, cmap="gray")
plt.title("Otsu Threshold")
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/content/download11.png")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

img_gray
```

smoothing

```python
gauss = cv2.GaussianBlur(img_gray, (5,5), 0)
median = cv2.medianBlur(img_gray, 5)
mean = cv2.blur(img_gray, (5,5))

plt.imshow(gauss, cmap="gray")
plt.show()
plt.imshow(median, cmap="gray")
plt.show()
plt.imshow(mean, cmap="gray")
plt.show()
```

sharpening

```python
mawar = cv2.imread("/content/mawar.png")
mawar_gray = cv2.cvtColor(mawar,cv2.COLOR_BGR2GRAY)
```

```python
kernell = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
sharpened = cv2.filter2D(mawar_gray, -1, kernell)

plt.imshow(sharpened, cmap="gray")
```

```python
lowpass = cv2.GaussianBlur(mawar_gray, (5,5), 1)

plt.imshow(lowpass, cmap="gray")
```

```python
highpass = cv2.subtract(mawar_gray, lowpass)

plt.imshow(highpass, cmap="gray")
```

```python
sharpen = cv2.add(mawar_gray, highpass)

plt.imshow(sharpen, cmap="gray")
```

segmentation

```python
# Global threshold
_, th_global = cv2.threshold(mawar_gray, 70, 255, cv2.THRESH_BINARY)

# Adaptive threshold (Gaussian)
th_adapt = cv2.adaptiveThreshold(mawar_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                 cv2.THRESH_BINARY, 11, 2)

# Otsu threshold
_, th_otsu = cv2.threshold(mawar_gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(th_global, cmap="gray"); plt.title("Global Threshold"); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(th_adapt, cmap="gray"); plt.title("Adaptive Threshold"); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(th_otsu, cmap="gray"); plt.title("Otsu Threshold"); plt.axis("off")
plt.show()
```
